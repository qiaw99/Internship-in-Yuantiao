## 折线图以及曲线

```python
import matplotlib.pyplot as plt

# 创建画布，figsize指定图的长宽，dpi指定图像的清晰度
plt.figure(figsize=(10, 6), dpi=100)

plt.plot([1, 2, 3, 4, 5, 6, 7], [10, 15, 13, 18, 16, 20, 10])

plt.show()
```

```python
import matplotlib.pyplot as plt
import random

plt.figure(figsize=(10, 6), dpi=100)

x = range(0, 60)
y = [random.uniform(15, 18) for i in x]

plt.plot(x, y)

# 设置x，y轴刻度
x_ticks_label = ["11:{}".format(i) for i in x]
y_ticks_label = range(40)

# 修改x，y轴坐标刻度显示
plt.xticks(x[::5], x_ticks_label[:: 5])
plt.yticks(y_ticks_label[::5])

plt.grid(True, linestyle = "--", alpha=0.5)

plt.xlabel("Time")
plt.ylabel("Temperature")

plt.title("11:00-12:00 temperature change", fontsize=20)

# plt.savefig("test.png")

plt.show()
```

```python
import matplotlib.pyplot as plt
import random

plt.figure(figsize=(10, 6), dpi=100)

x = range(0, 60)
y = [random.uniform(15, 18) for i in x]
y2 = [random.uniform(5, 10) for i in x]

plt.plot(x, y, label="shanghai")
# -实线 --虚线 -.点划线 :点虚线 ''留空、空格
plt.plot(x, y2, color='r', linestyle="--", label="beijing")

# 显示图例
plt.legend(loc="best")

# 设置x，y轴刻度
x_ticks_label = ["11:{}".format(i) for i in x]
y_ticks_label = range(40)

# 修改x，y轴坐标刻度显示
plt.xticks(x[::5], x_ticks_label[:: 5])
plt.yticks(y_ticks_label[::5])

plt.grid(True, linestyle = "--", alpha=0.5)

plt.xlabel("Time")
plt.ylabel("Temperature")

plt.title("11:00-12:00 temperature change", fontsize=20)

# plt.savefig("test.png")

plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = range(0, 60)
y = [random.uniform(15, 18) for i in x]
y2 = [random.uniform(5, 10) for i in x]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=100)

# 绘制图像
axes[0].plot(x, y, label="shanghai")
axes[1].plot(x, y2, color='r', linestyle="--", label="beijing")

# 显示图例
axes[0].legend(loc="best")
axes[1].legend(loc="best")

# 设置x，y轴刻度
x_ticks_label = ["11:{}".format(i) for i in x]
y_ticks_label = range(40)

# 修改x，y轴坐标刻度显示
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_ticks_label[::5])
axes[0].set_yticklabels(y_ticks_label[::5])

axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_ticks_label[::5])
axes[1].set_yticklabels(y_ticks_label[::5])

axes[0].grid(True, linestyle = "--", alpha=0.5)
axes[1].grid(True, linestyle = "--", alpha=0.5)

axes[0].set_xlabel("time")
axes[0].set_ylabel("temperature")
axes[1].set_xlabel("time")
axes[1].set_ylabel("temperature")

axes[0].legend(loc=1)
axes[1].legend(loc=1)

plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
y = np.sin(x)

plt.figure(figsize=(20, 8), dpi=100)

plt.plot(x, y)

plt.grid(linestyle='-.')

plt.show()
```

## 散点图： scatter(x, y)


```python
x = [random.uniform(0, 10) for i in range(10)]
y = [random.uniform(0, 10) for i in range(10)]

plt.figure(figsize=(20, 8), dpi = 100)

plt.scatter(x, y)

plt.show()
```

## 柱状图

```python
name = range(11)
x = range(11)
y = [67384, 32912, 73292, 92012,41234, 6759, 5931, 12359, 50293, 8302, 42098]

plt.figure(figsize=(20, 8), dpi = 100)

plt.bar(x, y, width=0.5, color=['b','r','g','y','c','m','y','k','c','g','b'])

# 修改x轴显示
plt.xticks(x, name)

plt.grid(linestyle='--', alpha=0.5)

plt.title("name comparision")

plt.show()
```
