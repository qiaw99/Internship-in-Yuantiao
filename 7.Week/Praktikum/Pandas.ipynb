## 缺失值处理
```python
import pandas as pd
import numpy as np

# 显示数据是否为NaN
pd.notnull(data) 

# 判断所有数据是否都不为NaN
np.all(pd.notnull(data))

# 判断有数据是NaN
np.any(pd.isnull(data))  
# 去除NaN的数据
data.dropna()

# 替换NaN 
data["column"].fillna(value, inplace=True)

for i in data.columns: 
    if np.any(pd.isnull(data[i])) == True:
        print(i)
        data[i].fillna(data[i].mean(), inplace = True)
        
# 把其他字符串替换成NaN, to_replace是替换前的值，value是替换后的值  
data.replace(to_replace = "", value = np.nan)
```

### pd.merge(left, right, how="inner", on=None)
- 可以指定按照两组数据的共同键值对合并或者左右各自连接
- on指定共同键
- how按照什么方式连接

| Merge Method |  SQL Join Name   |                Description                |
| ------------ | -------------- | --------------------------------------- |
| left         | LEFT OUTER JOIN  |       Use keys from left frame only       |
| right        | RIGHT OUTER JOIN |      Use keys from right frame only       |
| outer        | FULL OUTER JOIN  |    Use union of keys from both frames     |
| inner        |    INNER JOIN    | Use intersection of keys from both frames |
